```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
df.sample()
```

# Standardization using sklearn

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df[['Age','EstimatedSalary']])

# age_scaled = scaler.transform(df[['Age']])
# salary_scaled = scaler.transform(df[['EstimatedSalary']])
scaled_data = scaler.transform(df[['Age', 'EstimatedSalary']])
```

```python
df_scaled = pd.DataFrame(scaled_data, columns=['Age', 'EstimatedSalary'])
df_scaled.head()
```

```python
df_scaled.mean()
```

```python
df_scaled.std()
```

# Visulization

```python
sns.pairplot(df_scaled)
```

```python
plt.figure(figsize=(15, 15))

plt.subplot(2, 2, 1)  # Changed to (2, 1, 1) for the first subplot
sns.scatterplot(x='Age', y='EstimatedSalary', data=df)
plt.title(f"Original Data")  # Title might not be relevant here
plt.legend()

plt.subplot(2, 2, 2)  # Changed to (2, 1, 2) for the second subplot
sns.scatterplot(x='Age', y='EstimatedSalary', data=df_scaled)
plt.title(f"Scaled Data")  # Title might not be relevant here
plt.legend()

plt.tight_layout()
plt.show()
```
