# Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import dataset

```python
df = pd.read_csv('titanic_toy.csv')
df.head()
```

# Check missing values

```python
df.isna().sum()
```

```python
df.isnull().mean()*100
```

# Fill with mean and median

```python
df['Age'].mean(), df['Age'].median()
```

```python
df['Age_mean'] = df['Age'].fillna(df['Age'].mean())
df['Age_median'] = df['Age'].fillna(df['Age'].median())
```

```python
df['Fare'].mean(), df['Fare'].median()
```

```python
df['Fare_mean'] = df['Fare'].fillna(df['Fare'].mean())
df['Fare_median'] = df['Fare'].fillna(df['Fare'].median())
```

```python
df.head()
```

# See Distribution

```python
figure, axes = plt.subplots(2, 2, figsize=(15, 10))
sns.histplot(df['Age'], ax=axes[0, 0], color='blue', label='Age')
sns.histplot(df['Age_mean'], ax=axes[0, 0], color='red', label='Age_mean')
sns.histplot(df['Age_median'], ax=axes[0, 0], color='green', label='Age_median')
axes[0, 0].legend()
axes[0, 0].set_title('Age Distribution Histogram')

sns.kdeplot(df['Age'], ax=axes[0, 1], color='blue', label='Age')
sns.kdeplot(df['Age_mean'], ax=axes[0, 1], color='red', label='Age_mean')
sns.kdeplot(df['Age_median'], ax=axes[0, 1], color='green', label='Age_median')
axes[0, 1].legend()
axes[0, 1].set_title('Age Distribution KDE')

sns.histplot(df['Fare'], ax=axes[1, 0], color='blue', label='Fare')
sns.histplot(df['Fare_mean'], ax=axes[1, 0], color='red', label='Fare_mean')
sns.histplot(df['Fare_median'], ax=axes[1, 0], color='green', label='Fare_median')
axes[1, 0].legend()
axes[1, 0].set_title('Fare Distribution Histogram')

sns.kdeplot(df['Fare'], ax=axes[1, 1], color='blue', label='Fare')
sns.kdeplot(df['Fare_mean'], ax=axes[1, 1], color='red', label='Fare_mean')
sns.kdeplot(df['Fare_median'], ax=axes[1, 1], color='green', label='Fare_median')
axes[1, 1].legend()
axes[1, 1].set_title('Fare Distribution KDE')

plt.tight_layout()
plt.show()
```
