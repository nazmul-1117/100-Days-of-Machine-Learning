# Combine into a dictionary

```python
def create_combined_dict(col_list, response_list_dict):
  """
  Creates a single dictionary from a list of dictionaries,
  extracting only the specified keys.

  Args:
    col_list: A list of keys to extract from each dictionary.
    response_list_dict: A list of dictionaries.

  Returns:
    A dictionary where keys are from col_list and values are lists
    containing the corresponding values from each dictionary in
    response_list_dict.
  """
  combined_dict = {}
  for key in col_list:
    combined_dict[key] = []

  for x in response_list_dict:
    for key in col_list:
      if key in x:
        combined_dict[key].append(x[key])

  return combined_dict

combined_dict = create_combined_dict(col_list, response_list_dict)
print(combined_dict)
```

```python
import random

def generate_random_barcodes(num_barcodes=30, barcode_length=13):
  barcodes = []
  for _ in range(num_barcodes):
    barcode = ''.join(random.choices('0123456789', k=barcode_length))
    barcodes.append(barcode)
  return barcodes

# Generate a list of 30 random barcodes
random_barcodes = generate_random_barcodes()

# Print the generated barcodes
for barcode in random_barcodes:
  print(barcode)
```
