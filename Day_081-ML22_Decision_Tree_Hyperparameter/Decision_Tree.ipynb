# Start a Journey with Decision Tree

# Import Libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")

import time
```

```python
from sklearn.datasets import load_iris, make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
```

# Create/Import datasets

```python
start_time_dataset = time.perf_counter()
```

```python
X, y = make_classification(n_samples=100000, n_classes=2, n_features=1000, random_state=1)
```

```python
X.shape, y.shape
```

```python
end_time_dataset = time.perf_counter()
execution_time_dataset = end_time_dataset - start_time_dataset
print(f"Datasets Creation time: {execution_time_dataset} seconds")
```

# Plot Datasets

```python
sns.scatterplot(x=X[:,0], y=X[:,1], hue=y)
```

# Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

# Decision Tree Algorithm Apply

## Model 1

```python
start_time = time.perf_counter()
```

```python
pred1_start_time = time.perf_counter()
dtc1 = DecisionTreeClassifier()
dtc1.fit(X_train, y_train)
accuracy_score(y_test, dtc1.predict(X_test))
```

```python
pred1_end_time = time.perf_counter()
pred1_execution_time = pred1_end_time - start_time
print(f"Prediction time: {pred1_execution_time} seconds")
```

## Model 2

```python
pred2_start_time = time.perf_counter()
dtc2 = DecisionTreeClassifier(max_depth=3)
dtc2.fit(X_train, y_train)
accuracy_score(y_test, dtc2.predict(X_test))
```

```python
pred2_end_time = time.perf_counter()
pred2_execution_time = pred2_end_time - pred2_start_time
print(f"Prediction time: {pred2_execution_time} seconds")
```

## Model 3

```python
pred3_start_time = time.perf_counter()
dtc3 = DecisionTreeClassifier(max_depth=3, criterion="entropy")
dtc3.fit(X_train, y_train)
accuracy_score(y_test, dtc3.predict(X_test))
```

```python
pred3_end_time = time.perf_counter()
pred3_execution_time = pred3_end_time - pred3_start_time
print(f"Prediction time: {pred3_execution_time} seconds")
```

## Model 4

```python
pred4_start_time = time.perf_counter()
dtc4 = DecisionTreeClassifier(max_depth=3, criterion="entropy", min_samples_split=10)
dtc4.fit(X_train, y_train)
accuracy_score(y_test, dtc4.predict(X_test))
```

```python
pred4_end_time = time.perf_counter()
pred4_execution_time = pred4_end_time - pred4_start_time
print(f"Prediction time: {pred4_execution_time} seconds")
```

## Model 5

```python
pred5_start_time = time.perf_counter()
dtc5 = DecisionTreeClassifier(max_depth=3, criterion="entropy", min_samples_split=10, min_samples_leaf=1)
dtc5.fit(X_train, y_train)
accuracy_score(y_test, dtc5.predict(X_test))
```

```python
pred5_end_time = time.perf_counter()
pred5_execution_time = pred5_end_time - pred5_start_time
print(f"Prediction time: {pred5_execution_time} seconds")
```

## Model 6

```python
pred6_start_time = time.perf_counter()
dtc6 = DecisionTreeClassifier(max_depth=3, criterion="entropy", min_samples_split=10,
                              min_samples_leaf=1, min_impurity_decrease=0.01)
dtc6.fit(X_train, y_train)
accuracy_score(y_test, dtc6.predict(X_test))
```

```python
pred6_end_time = time.perf_counter()
pred6_execution_time = pred6_end_time - pred6_start_time
print(f"Prediction time: {pred6_execution_time} seconds")
```

## Model 7

```python
pred7_start_time = time.perf_counter()
dtc7 = DecisionTreeClassifier(max_depth=3, criterion="gini", min_samples_split=50,
                              min_samples_leaf=1, min_impurity_decrease=0.01,
                              splitter="random")
dtc7.fit(X_train, y_train)
accuracy_score(y_test, dtc7.predict(X_test))
```

```python
pred7_end_time = time.perf_counter()
pred7_execution_time = pred7_end_time - pred7_start_time
print(f"Prediction time: {pred7_execution_time} seconds")
```

## Model 8

```python
pred8_start_time = time.perf_counter()
dtc8 = DecisionTreeClassifier(max_depth=3, criterion="gini", min_samples_split=50,
                              min_samples_leaf=1, min_impurity_decrease=0.01,
                              splitter="random", max_features=50)
dtc8.fit(X_train, y_train)
accuracy_score(y_test, dtc8.predict(X_test))
```

```python
pred8_end_time = time.perf_counter()
pred8_execution_time = pred8_end_time - pred8_start_time
print(f"Prediction time: {pred8_execution_time} seconds")
```

```python
pred_end_time = time.perf_counter()
pred_execution_time = pred_end_time - start_time
print(f"Total Prediction time: {pred_execution_time} seconds")
```
