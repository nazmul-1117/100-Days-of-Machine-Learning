```
Categorical Data basically Two Tyeps
1. Count Plot -general purpose
2. Pie-Chart - % calculation
```

## Season-Wise

````python
plt.figure(figsize=(20, 20))

plt.subplot(4, 3, 1)  # Create subplot in a 3x3 grid
sns.countplot(data=df, x='season', color='green', label='Count Plot')  # Red distribution
plt.title(f"Distribution of season")
plt.legend()

plt.subplot(4, 3, 2)
df['season'].value_counts().plot(kind='pie', autopct='%1.1f%%', label='Pie Chart', figsize=(25, 20))
plt.title(f"Distribution of season")
plt.legend()

plt.tight_layout()
plt.show()
````
