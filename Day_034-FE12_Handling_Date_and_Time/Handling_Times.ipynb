# Convert to datetime

```python
df['time'] = pd.to_datetime(df['date'])
```

#Extract `houres`, `minutes` and `second`

```python
df['houres'] = df['time'].dt.hour
df['minutes'] = df['time'].dt.minute
df['second'] = df['time'].dt.second
```

```python
df.drop('msg', axis=1, inplace=True)
```

```python
df['tm'] = df['time'].dt.time
```
